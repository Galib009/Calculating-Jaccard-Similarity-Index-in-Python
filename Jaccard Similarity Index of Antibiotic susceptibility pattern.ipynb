```python
#!/usr/bin/env python
import pandas as pd
 
from math import*

# defining a user defined function for calculating Jaccard similarity - 

def jaccard_similarity(x,y):
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)

# checking if it works -
print (jaccard_similarity([0,1,2,5,6],[0,2,3,5,7,9]))
```

```python
### Child stool
C = pd.read_excel('AST_for_jaccard.xls', sheet_name= 'CS', index_col = False, keep_default_na=False, na_values=[''] )
for k in C.columns[1:] : 
    i = C['Drug_name'] + '_' + C['{}'.format(k)]
    i.to_csv('CS_P/{}.txt'.format(k), header = False, index = False)
```

```python
### Mother Stool
M = pd.read_excel('AST_for_jaccard.xls', sheet_name= 'MS', index_col = False, keep_default_na=False, na_values=[''] )
for k in M.columns[1:] :
    i = M['Drug_name'] + '_' + M['{}'.format(k)]
    i.to_csv('MS_P/{}.txt'.format(k), header = False, index = False)
```

```python
### Drinking water
W = pd.read_excel('AST_for_jaccard.xls', sheet_name= 'WU', index_col = False, keep_default_na=False, na_values=[''] )
for k in W.columns[1:] :
    i = W['Drug_name'] + '_' + W['{}'.format(k)]
    i.to_csv('WU_P/{}.txt'.format(k), header = False, index = False)
```

```python
# For calculating and saving in excel file the results:
# x, y and z values will get the jaccard values values of CSvsMS, CSvsWU and MSvsWU for 55 isolates - 
x= list(range(1,2))
y= list(range(1,2))
z= list(range(1,2))

# for loop will generate jaccard from files in the all_p folder and save it in x and y
for k in range(1,56):
    # .format(k) is used to take k as variable
    with open('All_P/CS_HH{} .txt'.format(k)) as f:
        HHC = f.read().splitlines()
    with open('All_P/MS_HH{} .txt'.format(k)) as f:
        HHM = f.read().splitlines()
    with open('All_P/WU_HH{} .txt'.format(k)) as f:
        HHW = f.read().splitlines()
    x[k-1] = jaccard_similarity(HHC,HHM)
    y[k-1] = jaccard_similarity(HHC,HHW)
    z[k-1] = jaccard_similarity(HHM,HHW)
    
    # increasing the value to keep the next value 
    x.append(0)
    y.append(0)
    z.append(0)
# print(x)
# print(y)
d= {'CSvsMS_55': x, 'CSvsWU_55':y,'MSvsWU_55':z}
df_55 = pd.DataFrame(d)
df_55
df_55.to_excel(r'jaccard_55_HH_T_plus_WU.xlsx', index = False)
```

```python
## to use the columns name of excel file and save as ROW : 
Co = pd.read_excel('excel_files/aaa.xls', index_col = False, keep_default_na=False, na_values=[''] )

# making a dictionary for taking this in dataframe in pandas:
d= {'Sample_Name' : Co.columns[1:], 'CSvsMS_55': x[:-1], 'CSvsWU_55':y[:-1],'MSvsWU_55':z[:-1]}
df_55 = pd.DataFrame(d)
df_55.to_excel(r'jaccard_55_HH_with_column_names_with_WU.xlsx', index = False)
```
